```python
from keras.applications.vgg16 import VGG16

model = VGG16()
model.summary()
```

```python
#library
import requests
import numpy as np
import cv2
import matplotlib.pyplot as plt
from PIL import Image

urls=[]
# mobil
urls.append("https://carmudi-journal.icarcdn.com/carmudi-id/wp-content/uploads/2022/02/17172020/Toyota-All-New-Voxy-Indonesia.jpg")
urls.append("https://4.bp.blogspot.com/-Pel1KXcz1P4/UaaKTDEvbcI/AAAAAAAAAj4/OrA8HbEepyU/s1600/Gambar+interior+mobil+tata+nano+Dan+Eksterior+mobil+ini+1.jpg")
# motor
urls.append("https://images.autofun.co.id/file1/9c64ee8a8393437199f0afff3f1e1702_1200.jpg")
urls.append("https://mansarpost.files.wordpress.com/2015/04/wpid-modifikasi-yamaha-vixion-nvl-kuning-bumble-bee-jpg.jpeg")

imgs=[]
plt.figure(figsize=(15,6))
for u in range(0,len(urls)):
  r = requests.get(urls[u], stream=True) 
  img = np.array(Image.open(r.raw)) 
  plt.subplot(1,len(urls),u+1)
  plt.imshow(img)
  img = cv2.resize(img, (224,224))
  imgs.append(img)
plt.show()
imgs = np.array(imgs)
print (imgs.shape)
```

```python
yh = model.predict(imgs)
for i in range(len(urls)):
  if(np.argmax(yh[i]) == 436  or np.argmax(yh[i]) == 565 or np.argmax(yh[i])
  print("Mobil")
  elif(np.argmax(yh[i]) == 670 or np.argmax(yh[i]) == 671):
    print("Motor")
  else:
     print("lainnya")
```
